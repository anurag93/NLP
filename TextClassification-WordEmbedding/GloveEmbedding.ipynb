```
from glove import Corpus, Glove
from nltk.corpus import stopwords 
import nltk
from nltk.stem import WordNetLemmatizer
import pandas as pd
import re
```

```
nltk.download('stopwords')
```

```
nltk.download('wordnet')
```

```
data = pd.read_csv(r'/content/data/Reviews.csv', nrows = 100000, usecols=['Summary','Text'])
```

```
lines = list()
for summary in data['Summary']:
  lines.append(summary)
```

```
for story in data['Text']:
  story_replaced = re.sub(r'\.+', '.', story)
  story_split = story_replaced.split(".")
  for line in story_split:
    lines.append(line)
```

```
len(lines)
```

```
stop_words=set(stopwords.words('english'))
```

```
lines_without_stopwords=[]
for line in lines:
  temp_line=[]
  for word in str(line).split():
    if word not in stop_words:
      temp_line.append(word)
  string = ' '
  lines_without_stopwords.append(string.join(temp_line))
  
lines = lines_without_stopwords
```

```
len(lines)
```

```
wordnet_lemmatizer = WordNetLemmatizer()
```

```
line_with_lemmas = []
for line in lines:
  temp_line = []
  for word in line.split():
    temp_line.append(wordnet_lemmatizer.lemmatize(word))
  string = ' '
  line_with_lemmas.append(string.join(temp_line))
lines = line_with_lemmas
```

```
new_lines = []
for line in lines:
  new_lines.append(line.split())
lines = new_lines
```

```
corpus = Corpus()
```

```
corpus.fit(lines, window=10)
```

```
glove = Glove(no_components=100, learning_rate=0.05)
glove.fit(corpus.matrix, epochs=30, no_threads=4, verbose=True)
```

```
glove.add_dictionary(corpus.dictionary)
```
